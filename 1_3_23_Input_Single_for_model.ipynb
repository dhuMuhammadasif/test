```python
import numpy as np
import math
t = np.linspace(0, 1, 2500)

t
```

```python
import matplotlib.pyplot as plt

f_x = []
for t_in in t:
    v_t = 3.75*math.exp(-0.9 * t_in)*(math.sin(2*math.pi*3*t_in)+1)
    f_x.append(v_t)
train_x = np.array(f_x)
plt.plot(f_x)
plt.show()
```

```python
max_xp = max(f_x)
min_xp = min(f_x)

# max_xp = max(inc_x)
# min_xp = min(dec_x)

max_fxp = 1 - np.exp(-max_xp)
min_fxp = 1 - np.exp(min_xp)

print(max_xp, max_fxp)
print(min_xp, min_fxp)
```

```python
y = []
for i in range(1, len(f_x)):
    del_x = f_x[i] - f_x[i-1]
    if del_x > 0:
        # y_inc = (
        #     0.0228+0.0012*np.exp(-(train_x[i]-max_xp))) * (train_x[i]-max_xp)+0.1812*max_fxp
        y_inc = (0.00223-0.0012*np.exp(-(f_x[i]-max_xp)))+ 0.9812*max_fxp
        y.append(y_inc)
    else:
        y_dec = (0.00323+0.0011*np.exp((f_x[i]-min_xp))) - 0.0058*min_fxp
            # 0.0232+0.0012*np.exp((train_x[i]-min_xp))) * (train_x[i]-min_xp)+1.1258*min_fxp
        y.append(y_dec)

plt.figure(figsize=(11, 7))
plt.scatter(y,f_x[1:])
plt.show()
```
